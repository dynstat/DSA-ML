### Sorting

#### Bubble Sort
- compare with its right element
- swap if bigger 
- eventually the bigger will shift to the right and biggest to the right most
- Repeat from the start with leaving the rightmost element.

```python
# Bubble sort implementation 
def bubble_sort(arr):
    start=0
    l = len(arr)
    
    while l > start:
        for idx in range(l-1):
            if arr[idx] > arr[idx+1]:
                arr[idx] , arr[idx+1] = arr[idx+1], arr[idx] # swapping
        l -= 1

array = [23, 0, 3, 0, 56, 12, 9, 3,5,8,3,1,0]

print("Before:", array)
bubble_sort(array)
print("After: ",array)


# NOTE: Can be optimized by keep on tracking the last sorted index,so that 
# we don't have to compare each n every index. 
```

### Merge Sort
- split into two parts/functions.
- one function is for merging two sorted arrays of different sizes
- other is to split the arrays into two halfs (recursively) and using the first function inside it.


```python
# Merge sort implementation
def combine_arrays(arr1, arr2):
    i = j = 0
    l1 = len(arr1)
    l2 = len(arr2)
    
    M = []
    while i < l1 and j < l2:
        if arr1[i] <= arr2[j]:
            M.append(arr1[i])
            i += 1
        else:
            M.append(arr2[j])
            j +=1
    while i<l1:
        M.append(arr1[i])
        i += 1
                        
    while j<l2:
        M.append(arr2[j])
        j += 1
    return M

def merge_sort(arr):
    """ This will split the array into two parts recursively and 
    the join them using the combine_arrays() function"""
    if len(arr) <= 1 :
        return arr
    
    l = len(arr)
    mid = l // 2
    left_half = arr[:mid]
    right_half = arr[mid:]
    
    sorted_left_half = merge_sort(left_half)
    sorted_right_half = merge_sort(right_half)
    
    # merging/combining the supposidly, sorted arrays
    return combine_arrays(sorted_left_half,sorted_right_half)


array = [23, 0, 3, 0, 56, 12, 9, 3,5,8,3,1,0]

print("Before:", array)
ans = merge_sort(array)
print("After: ",array)
print("After: ",ans)
```
