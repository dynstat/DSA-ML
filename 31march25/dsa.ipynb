```python
## INSERTION SORT

def insertion_sort(arr):
    for orange in range(1, len(arr)):
        left = 0
        right = orange -1
        curr_val = arr[orange]
        
        ## Finding the correct position where we have to put the "curr_val" in the array to the left (previously sorted logical subarray)
        while left <= right :
            if curr_val < arr[right]:
                right -= 1
            else:
                break
        
        ## we got the location in the "right" variable as (right + 1) index, i.e the value jsut after the "right" value index
        # and need to shift the elements in order to make the space for curr_val element.
        
        ## shifting of the elements from "right" to "orange - 1", start from reverse direction (r --> l)
        t = orange  # temp variable so that i don't mess up the orange variable of the for loop
        while t > right+1:  # shifting has to be done for indexes from right+1 to the orange-1, all to the one index right of themselves.
            arr[t] = arr[t - 1]
            t -= 1
        
        # by now, the array items have been shifted accordingly
        # putting the "curr_val" at the correct location i.e. at "right" + 1 index
        
        arr[right + 1] = curr_val
             
        

array = [23, 1, 10, 5 , 2]
print("Given unsorted Array: ", array)
# calling the insertion sort function
insertion_sort(array)
print("Array after insertion sorting: ", array)
        
        
        
        
```

```python
# MERGE SORT
def merge_two_sorted_arrays(left_arr, right_arr):
    i = j = k = 0
    M = []
    while i < len(left_arr) and j < len(right_arr):
        if left_arr[i] < right_arr[j]:
            M.append(left_arr[i])
            i+=1 
        else:
            M.append(right_arr[j])
            j+=1
    
    while i < len(left_arr):
        M.append(left_arr[i])
        i+=1 
    while j < len(right_arr):
        M.append(right_arr[i])
        j+=1 
    return M

def merge_sort(arr):
    if len(arr) <= 1:
        return arr
    
    # divide the arr into two parts
    mid = len(arr)//2
    left_arr = merge_sort(arr[:mid])
    right_arr = merge_sort(arr[mid:])
    
    
    return merge_two_sorted_arrays(left_arr=left_arr, right_arr=right_arr)


### Main code for testing the function
array = [23, 1, 10, 5 , 2]
print("Given unsorted Array: ", array)
# calling the merge sort function
sorted_array = merge_sort(array)
print("Array after insertion sorting: ", sorted_array)
```

```python
def practical_merge_sort(arr):
    # Small buffer to improve performance
    BUFFER_SIZE = 32  # Adjust based on your needs
    buffer = [0] * BUFFER_SIZE
    
    def merge_sort_helper(start, end):
        # Use insertion sort for small subarrays
        if end - start <= 16:
            for i in range(start + 1, end):
                key = arr[i]
                j = i - 1
                while j >= start and arr[j] > key:
                    arr[j+1] = arr[j]
                    j -= 1
                arr[j+1] = key
            return
        
        mid = (start + end) // 2
        merge_sort_helper(start, mid)
        merge_sort_helper(mid, end)
        
        buffered_merge(start, mid, end)
    
    def buffered_merge(start, mid, end):
        if mid - start <= BUFFER_SIZE:
            # If left array fits in buffer, use it for efficient merging
            for i in range(start, mid):
                buffer[i-start] = arr[i]
            
            # Merge
            left = 0
            right = mid
            dest = start
            
            while left < mid - start and right < end:
                if buffer[left] <= arr[right]:
                    arr[dest] = buffer[left]
                    left += 1
                else:
                    arr[dest] = arr[right]
                    right += 1
                dest += 1
            
            # Copy remaining from buffer
            while left < mid - start:
                arr[dest] = buffer[left]
                left += 1
                dest += 1
        else:
            # Otherwise use in-place merge
            inplace_merge(start, mid, end)
    
    def inplace_merge(start, mid, end):
        # Use one of the in-place merge strategies mentioned above
        # This is a simplified version with rotations
        i = start
        j = mid
        
        while i < j and j < end:
            if arr[i] <= arr[j]:
                i += 1
            else:
                # Rotate to insert arr[j] at position i
                val = arr[j]
                # Shift elements
                for k in range(j, i, -1):
                    arr[k] = arr[k-1]
                arr[i] = val
                # Update pointers
                i += 1
                j += 1
                mid += 1
    
    merge_sort_helper(0, len(arr))
```

```python
### Main code for testing the function
array = [23, 1, 10, 5 , 2]
print("Given unsorted Array: ", array)
# calling the insertion sort function
practical_merge_sort(array)
print("Array after insertion sorting: ", array)
```
