```python
## INSERTION SORT

def insertion_sort(arr):
    for step in range(1, len(arr)):
        left = 0
        right = step -1
        curr_val = arr[step]
        
        ## Finding the correct position where we have to put the "curr_val" in the array to the left (previously sorted logical subarray)
        while left <= right :
            if curr_val < arr[right]:
                right -= 1
            else:
                break
        
        ## we got the location in the "right" variable as (right + 1) index, i.e the value jsut after the "right" value index
        # and need to shift the elements in order to make the space for curr_val element.
        
        ## shifting of the elements from "right" to "step - 1", start from reverse direction (r --> l)
        t = step  # temp variable so the i don't mess up the step variable of the for loop
        while t > right+1:  # shifting has to be done for indexes from right+1 to the step-1, all to the one step right of themselves.
            arr[t] = arr[t - 1]
            t -= 1
        
        # by now, the array items have been shifted accordingly
        # putting the "curr_val" at the correct location i.e. at "right" + 1 index
        
        arr[right + 1] = curr_val
             
        

array = [23, 1, 10, 5 , 2]
print("Given unsorted Array: ", array)
# calling the insertion sort function
insertion_sort(array)
print("Array after insertion sorting: ", array)
        
        
        
        
```

```python
# MERGE SORT
def merge_two_sorted_arrays(left_arr, right_arr):
    i = j = k = 0
    M = []
    while i < len(left_arr) and j < len(right_arr):
        if left_arr[i] < right_arr[j]:
            M.append(left_arr[i])
            i+=1 
        else:
            M.append(right_arr[j])
            j+=1
    
    while i < len(left_arr):
        M.append(left_arr[i])
        i+=1 
    while j < len(right_arr):
        M.append(right_arr[i])
        j+=1 
    return M

def merge_sort(arr):
    if len(arr) <= 1:
        return arr
    
    # divide the arr into two parts
    mid = len(arr)//2
    left_arr = merge_sort(arr[:mid])
    right_arr = merge_sort(arr[mid:])
    
    
    return merge_two_sorted_arrays(left_arr=left_arr, right_arr=right_arr)


### Main code for testing the function
array = [23, 1, 10, 5 , 2]
print("Given unsorted Array: ", array)
# calling the insertion sort function
sorted_array = merge_sort(array)
print("Array after insertion sorting: ", sorted_array)
```
